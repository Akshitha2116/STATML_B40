```python
#QUESTION-1:
import pandas as pd
from sklearn.preprocessing import minmax_scale
from sklearn.model_selection import train_test_split
#STEP 1-READ THE DATA:
data=pd.read_csv('train.csv')
#ASSUMING THE TARGET VARIABLE IS THE LAST COLUMN
features=data.iloc[:,:-1]
target=data.iloc[:,:-1]
#STEP 2-NORMALIZE THE DATA
scaler=minmax_scale(features)
#STEP 3-SPLIT THE DATA INTO TRAIN AND TEST
X_train,X_test,y_train,y_test=train_test_split(scaler,target,test_size=0.2,random_state=42)
#OUTPUT SHAPES TO VERIFY
print(f"Training feature shape: {X_train.shape}")
print(f"Training target shape: {y_train.shape}")
print(f"Testing feature shape: {X_test.shape}")
print(f"Testing target shape: {y_test.shape}")
```

```python
#QUESTION-2:
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
#STEP 1-READ THE DATA AND DESCRIBE IT:
data=pd.read_csv('train.csv')
print("Data Description")
print(data.describe())
#STEP2:FIND THE DATA TYPE AND SHAPE OF EACH COLUMN
print("\nData Types and Shape Of Each Column:")
print(data.dtypes)
print("Shape of Data:",data.shape)
#STEP3:FIND THE TARGET AND FEATURES
print("\nColumns in Data:")
print(data.columns)
target_column=data.columns[0]
X=data.drop(columns=[target_column])
Y=data[target_column]
print("\nfeatures and target:")
print("features(X)",X.head())
print("target(Y)",Y.head())
#STEP 4:FIND NULL VALUES AND FILL THEM
print("\nNull Values before filling:")
print(data.isnull().sum())
data.fillna(0,inplace=True)
print("\nNull Values after filling:")
print(data.isnull().sum())
X=data.drop(columns=[target_column])
Y=data[target_column]
#STEP 5:NORMALIZE ALL THE FEATURES
scaler=StandardScaler()
X_normalized=scaler.fit_transform(X)
#STEP 6:SPLIT THE DATA INTO TRAIN AND TEST
X_train,X_test,y_train,y_test=train_test_split(X_normalized,Y,test_size=0.2,random_state=42)
print("\ndata splitting:")
print("Training set size:",X_train.shape)
print("Testing set size:",X_test.shape)
```

```python
import pandas as pd
#LOAD YOUR EXCEL DATA
file_path='salary.xls'  #Update with your file path
data=pd.read_excel(file_path)
#DISPLAY THE FIRST FEW ROWS OF THE DATASET
print(data.head())
#ASSUMING THE LASR COLUMN IS THE
x=data.iloc[:,:-1]  #FEATURES
y=data.iloc[:,-1]   #TARGET
#DISPLAY THE FEATURES AND TARGET
print("Features:\n",x.head())
print("Target:\n",y.head())
import matplotlib.pyplot as plt
for column in x.columns:
    #CHNAGED X TO X
    plt.figure(figsize=(8,6))
    plt.scatter(x[column],y,alpha=0.5)
    plt.xlabel(column)
    plt.ylabel('Target')
    plt.show()
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_normalized=scaler.fit_transform(x)
y_normalized=scaler.fit_transform(y.values.reshape(-1,1))
#CONVERT THE BACK TO DATAFRAME FOR EASIER HANDLING
X_normalized=pd.DataFrame(x_normalized,columns=x.columns)
Y_normalized=pd.Series(y_normalized.flatten(),name='Target')
#DISPLAY THE NORMALIZED DATA
print("Normalized Features:\n",X_normalized.head())
print("Normalized Target:\n",Y_normalized.head())
```
